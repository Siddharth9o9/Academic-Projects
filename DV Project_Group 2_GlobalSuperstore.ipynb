# What is the region wise sales of the store

```python
#Give the sales of the store region wise in lakhs
grocerydata.groupby('Orders_Region').Sales.sum()/10**5
```

```python
grocerydata.groupby('Orders_Region').Sales.sum()/10**5
grocerydata_pivot=pd.pivot_table(grocerydata, values="Sales",columns="Orders_Region", aggfunc=np.sum)
ax=grocerydata_pivot.plot(kind="bar")
fig=ax.get_figure()
fig.set_size_inches(20,10)
ax.set_xlabel("Orders_Region")
ax.set_ylabel("sales")

for bar in ax.patches: 
    
  # Using Matplotlib's annotate function and 
  # passing the coordinates where the annotation shall be done 
  # x-coordinate: bar.get_x() + bar.get_width() / 2 
  # y-coordinate: bar.get_height() 
  # ha and va stand for the horizontal and vertical alignment 
    ax.annotate(format(bar.get_height(), '.2f'),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=10, xytext=(0, 8), 
                   textcoords='offset points') 
plt.show()
#Analysis shows Western Europe has the highest sales
```

# How much is the sales in differents  markets?

```python
grocerydata.groupby('Market').Sales.sum()
grocerydata_pivot=pd.pivot_table(grocerydata, values="Sales",columns="Market", aggfunc=np.sum)
ax=grocerydata_pivot.plot(kind="bar")
fig=ax.get_figure()
fig.set_size_inches(7,6)
ax.set_xlabel("Market")
ax.set_ylabel("Sales")
for bar in ax.patches: 
    
     ax.annotate(format(bar.get_height()/10**6, '.2f'),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=8, xytext=(0, 8), 
                   textcoords='offset points') 
plt.show()
```

# Interactive plot using Bokeh

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
```

```python
grocerydata.columns
```

```python
type(grocerydata)
```

```python
grocerydata.rename(columns = {'Customer Name':'CustomerName'}, inplace = True)
```

```python
grocerydata.rename(columns = {'Product Name':'ProductName'}, inplace = True)
```

```python
grocerydata.rename(columns = {'Order ID':'OrderID'}, inplace = True) 
```

```python
output_file('columndatasource_example.html')

sample = grocerydata
source = ColumnDataSource(sample)

p = figure(plot_height=600, plot_width=1000)

p.circle(x='Quantity', y='Sales',
         source=source,
         size=10, color='green')

p.title.text = 'ORDER QUANTITY AND SALES'
p.xaxis.axis_label = 'Quantity'
p.yaxis.axis_label = 'Sales'
```

```python
   
hover = HoverTool(tooltips=[('OrderID', '@OrderID'),('Sales', '@Sales'),('ProductName', '@ProductName'),('CustomerName', '@CustomerName'),('Quantity', '@Quantity')])

p.add_tools(hover)

show(p)
```
